## Initial Libs and First look at the data
- dataset looks clean
---

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('customers.csv')
print(df.info())
df.head()
```

## Proprocessing

---

- droping customerID
- renaming field for better access
- changing types to float 64

```python
df.drop('CustomerID', axis='columns', inplace=True)

df.rename(
    {
        'Annual Income (k$)' : 'income',
        'Spending Score (1-100)' : 'score'
    },
    axis='columns',
    inplace=True
)
df.rename(str.lower, axis='columns', inplace=True)

df[['age', 'income', 'score']] = df[['age', 'income', 'score']].astype('float64')

df.head()
```

- encoding gender into numeric

```python
from sklearn.preprocessing import LabelEncoder

df.gender = LabelEncoder().fit(['Female', 'Male']).transform(df.values[:, 0])
df.head()
```

- normalizing features

```python
from sklearn.preprocessing import MinMaxScaler

X = df.values
X_scaled = MinMaxScaler().fit_transform(X)
X[:5], X_scaled[:5]
```

## Clustering 
---
- we will cluster our customers by 3 different algorithms
  1. K-means
  2. Agglomerative Hierarchicy
  3. DBSCAN

## K-means
---
- first we have to find optimal K, this can be done by elbow method down below
- I think 4 is the k where the plot decreasing much slower

```python
from sklearn.cluster import KMeans 

def calculate_WCSS(points, kmax):
    WCSS = np.zeros(kmax+1)
    
    for k in range(1, kmax+1):
        kmeans_model = KMeans(init = "k-means++", n_clusters = k, n_init = 12).fit(points)
        WCSS[k] = kmeans_model.inertia_
    
    return WCSS

wcss = calculate_WCSS(X_scaled, 10)
plt.plot(wcss)
```

- labeling data with k=4

```python
k=4
kmeans_model = KMeans(init = "k-means++", n_clusters = k, n_init = 12).fit(X)

df['cluster'] = kmeans_model.labels_
df.head()
```

- as we can see, mean of gender in each cluster is near to 0.5, which means gender didn't affect our clusters
- our data is clustered by features age, income and score

```python
df.groupby('cluster').describe().T
```

```python
centroids = kmeans_model.cluster_centers_[:, 1:]
centroids
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection='3d')

colors = plt.cm.Spectral(np.linspace(0, 1, k))

for cluster, color in enumerate(colors):
    sub_df = df[df.cluster == cluster]
    
    x, y, z = sub_df[['age', 'income', 'score']].T.to_numpy()
    
    ax.scatter(x, y, z, color=color)
    ax.scatter(*centroids[cluster], color=color, edgecolor='k', s=50)
    
ax.set_xlabel('age')
ax.set_ylabel('income')
ax.set_zlabel('score')
```

```python
fig = plt.figure(figsize=(10, 10))

colors = plt.cm.Spectral(np.linspace(0, 1, k))

for cluster, color in enumerate(colors):
    sub_df = df[df.cluster == cluster]
    
    age, income, score = sub_df[['age', 'income', 'score']].T.to_numpy()

    plt.subplot(2, 2, 1, xlabel='age', ylabel='income')
    plt.scatter(age, income, color=color)
    plt.scatter(*centroids[cluster][[0, 1]], color = color ,edgecolors='k', s=30)

    plt.subplot(2, 2, 2, xlabel='age', ylabel='score')
    plt.scatter(age, score, color=color)
    plt.scatter(*centroids[cluster][[0, 2]], color = color ,edgecolors='k', s=30)

    
    plt.subplot(2, 2, 3, xlabel='income', ylabel='score')
    plt.scatter(income, score, color=color)
    plt.scatter(*centroids[cluster][[1, 2]], color = color ,edgecolors='k', s=30)
```

## Hierarchical / Agglomerative
---
- in k-means we find out that 4 clusters are enough, so we are doing Hierarchical with 4 clusters.

```python
from sklearn.cluster import AgglomerativeClustering 

clusters = 4
agglom_model = AgglomerativeClustering(n_clusters = 5, linkage = 'complete').fit(X_scaled)

df['cluster'] = agglom_model.labels_
df.head()
```

- now we have our model and labeled data we have to plot our dendrogram
- dendro gram requires Z wich is linkage matrix and needs to be calculated with distance matrix


```python
from sklearn.metrics.pairwise import euclidean_distances
from scipy.cluster import hierarchy
from scipy.spatial.distance import squareform

dist_matrix = euclidean_distances(X_scaled, X_scaled) 

Z = hierarchy.linkage(dist_matrix, 'complete')

Z[:5]
```

```python
fig = plt.figure(figsize=(18,50))


dendro = hierarchy.dendrogram(
    Z,
    leaf_label_func= lambda i: f" { df['age'][i] } - { df['income'][i] } - { df['score'][i] } ",
    color_threshold=0.5*max(Z[:,2]),
    leaf_font_size=12,
    orientation='right',
    above_threshold_color='gray',
)
```

## DBSCAN
---
- finding optimal eps and min_samples are hard so I skipped this part by playing with data to find best

```python
from sklearn.cluster import DBSCAN

dbscan_model = DBSCAN(eps=0.2, min_samples=6).fit(X_scaled)

df['cluster'] =  dbscan_model.labels_

df.head()
```

- wow! this time can say our clusters are seprated by gender too

```python
df.groupby('cluster').describe().T
```

```python
females, males = df[df.gender == 0], df[df.gender == 1]

females_labels = females['cluster'].unique()
males_labels = males['cluster'].unique()
females_labels, males_labels
```

```python
fig = plt.figure(figsize=(10, 10))
plt.suptitle('Clustered Scatter Plots for Females', fontsize=16)

colors = plt.cm.viridis(np.linspace(0, 1, len(females_labels)))

for cluster, color in zip(females_labels, colors):
    color = 'k' if cluster == -1 else color
    
    sub_df = females[females.cluster == cluster]
    
    age, income, score = sub_df[['age', 'income', 'score']].T.to_numpy()

    plt.subplot(2, 2, 1, xlabel='age', ylabel='income')
    plt.scatter(age, income, color=color)

    plt.subplot(2, 2, 2, xlabel='age', ylabel='score')
    plt.scatter(age, score, color=color)

    plt.subplot(2, 2, 3, xlabel='income', ylabel='score')
    plt.scatter(income, score, color=color)
```
